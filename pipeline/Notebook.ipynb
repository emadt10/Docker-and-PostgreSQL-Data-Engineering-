```python
import pandas as pd

# Read a sample of the data
prefix = 'https://github.com/DataTalksClub/nyc-tlc-data/releases/download/yellow/'
url = f'{prefix}/yellow_tripdata_2021-01.csv.gz'
df = pd.read_csv(url, nrows=100)

# Display first rows
df.head()

# Check data types
df.dtypes

# Check data shape
df.shape
```

```python
dtype = {
    "VendorID": "Int64",
    "passenger_count": "Int64",
    "trip_distance": "float64",
    "RatecodeID": "Int64",
    "store_and_fwd_flag": "string",
    "PULocationID": "Int64",
    "DOLocationID": "Int64",
    "payment_type": "Int64",
    "fare_amount": "float64",
    "extra": "float64",
    "mta_tax": "float64",
    "tip_amount": "float64",
    "tolls_amount": "float64",
    "improvement_surcharge": "float64",
    "total_amount": "float64",
    "congestion_surcharge": "float64"
}

parse_dates = [
    "tpep_pickup_datetime",
    "tpep_dropoff_datetime"
]

df = pd.read_csv(
    url,
    nrows=100,
    dtype=dtype,
    parse_dates=parse_dates
)
```

```python
from sqlalchemy import create_engine
engine = create_engine('postgresql://root:root@localhost:5432/ny_taxi')
```

```python
print(pd.io.sql.get_schema(df, name='yellow_taxi_data', con=engine))
```

```python
df.head(n=0).to_sql(name='yellow_taxi_data', con=engine, if_exists='replace')
```

```python
df_iter = pd.read_csv(
    url,
    dtype=dtype,
    parse_dates=parse_dates,
    chunksize=100_000
)
```

```python

from tqdm.auto import tqdm
first_chunk = next(df_iter)

# Insert first chunk
first_chunk.to_sql(
    name="yellow_taxi_data",
    con=engine,
    if_exists="append",
    index=False
)
print("Inserted first chunk:", len(first_chunk))

# ----------------------------
# 5. Insert remaining chunks with progress bar
# ----------------------------
for df_chunk in tqdm(df_iter, desc="Inserting chunks"):
    df_chunk.to_sql(
        name="yellow_taxi_data",
        con=engine,
        if_exists="append",
        index=False
    )
```

```python
# 1️⃣ Pehla chunk nikal lo
first_chunk = next(df_iter)

# ----------------------------
# 2️⃣ Table schema create karo (no data)
# ----------------------------
first_chunk.head(0).to_sql(
    name="yellow_taxi_data",
    con=engine,
    if_exists="replace",
    index=False
)
print("Table schema created.")

# ----------------------------
# 3️⃣ Pehla chunk insert karo
# ----------------------------
first_chunk.to_sql(
    name="yellow_taxi_data",
    con=engine,
    if_exists="append",
    index=False
)
print("Inserted first chunk:", len(first_chunk))

# ----------------------------
# 4️⃣ Baaki chunks insert karo (no tqdm)
# ----------------------------
for df_chunk in df_iter:
    df_chunk.to_sql(
        name="yellow_taxi_data",
        con=engine,
        if_exists="append",
        index=False
    )
    print("Inserted chunk:", len(df_chunk))

print("All chunks inserted successfully!")
```
